```python
# A bit of setup
import numpy as np
import matplotlib.pyplot as plt
 
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
 
# for auto-reloading extenrnal modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
np.random.seed(0)
N = 100 # number of points per class
D = 2 # dimensionality
K = 3 # number of classes
X = np.zeros((N*K,D)) #(300,2)
y = np.zeros(N*K, dtype='uint8') #(300,1)
for j in range(K):
  ix = range(N*j,N*(j+1))
  r = np.linspace(0.0,1,N) # radius
  t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # theta
  X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
  y[ix] = j
fig = plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y,edgecolors="k", s=40, cmap=plt.cm.Spectral)
plt.xlim([-1,1])
plt.ylim([-1,1])
#fig.savefig('spiral_raw.png')
```

```python
#input yha 2 hoga (size 2)(2 point x1, x2)
#output hoga 3  (size 3)
#hidden layer me 100 neurons
```

```python
h=100 #SIZE OF HIDDEN LAYER 
W=0.01*np.random.randn(D,h) #(2,100)
b= np.zeros((1,h)) #(1,100)
 
W2=0.01*np.random.randn(h,K) #(100,3)
b2=np.zeros((1,K))#(1,3)
 
step_size=1e-0 #(floating poin bnane  k liye) #learning_rate 
reg=1e-3 #regularization strengh
 
 
m=X.shape[0] #300 bcz 300 example h yha
 
 
for i in range(10000):
    #forward_propagation
    
    #(1) compute the raw output
    #hidden_layer_output=np.dot(X,W) +b  #np.dot(X,W).shape=(300,100) , b.shape=(100,1)
    hidden_layer_output=np.maximum(0,np.dot(X,W)+b) #(300,100) #linear fnction na bna rhe esliye , for making non linear,#reLU ,3graph teda-meda bnane ko 
    scores=np.dot(hidden_layer_output,W2)+b2 #(300,3)
    #(2)compute class props #softmax
    
    exp_score=np.exp(scores) #(300,3)
    prob=exp_score/np.sum(exp_score,axis=1,keepdims=True)
    
    
    #(3)calculate loss 
    
    correct_logprobs=-np.log(prob[range(m),y]) #(300,1)
    
    data_loss = np.sum(correct_logprobs)/m #scalar
    
    #4 regularization
    
    reg_loss=0.5*reg*np.sum(W*W) + 0.5*reg*np.sum(W2*W2)
    loss= data_loss + reg_loss  #loss minimize & prevent overfitting
    
    
    if i%1000==0:
        print(f"iter{i}:loss{loss }")
    
    #backproopagation
    
    #1(1 )find gradient of the loss(y^-y)
    dprobs =prob  #(300,3)
    
    dprobs[range(m),y]-=1  #(300,3)
    dprobs/=m #300,3
    
    
    #(2) backprob into W2 , b2
    dW2 = np.dot(hidden_layer_output.T,dprobs)  #T is to transpose matrics
    db2 = np.sum(dprobs,axis=0) #(1,3)
    
    #(3.1) for  backprob into hidden layerv,find gradient
    dhidden = np.dot(dprobs,W2.T) #(300,100)
    dhidden[hidden_layer_output <=0 ] =0 #backprop ReLU
    
    #(3.2) Backprop into W,b
    dW=np.dot(X.T,dhidden) #(2,100)
    db=np.sum(dhidden,axis=0 ) #(!1,100)
    
    #regularizATION
    dW2+=reg*W2
    dW+=reg*W
    
    
    #4 parameter update (Update rule)
    
    W-= step_size*dW
    b-=step_size*db
    W2-=step_size*dW2
    b2-= step_size*db2
```

```python
hidden_layer_output = np.maximum(0,np.dot(X,W)+b) #300,100
scores=np.dot (hidden_layer_output ,W2 )+b2
pred = np.argmax(scores ,axis=1)
```

```python
# plot the resulting classifier
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = np.dot(np.maximum(0, np.dot(np.c_[xx.ravel(), yy.ravel()], W) + b), W2) + b2
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)
fig = plt.figure()
plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
#fig.savefig('spiral_net.png')
```
